```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

creating a blank image

```python
blank_img=np.zeros(shape=(720,720,3),dtype=np.int32)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

Drawing a rectangle on blank image

```python
cv2.rectangle(blank_img,pt1=(590,10),pt2=(690,200),color=(0,225,0),thickness=10)
```

```python
plt.imshow(blank_img)
```

Drawing a square in the middel

```python
cv2.rectangle(blank_img,pt1=(300,300),pt2=(400,400),color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_img)
```

Drawing a circle on image

```python
cv2.circle(img=blank_img,center=(200,200),radius=10,color=(255,0,0),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(200,200),radius=50,color=(255,0,0),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(200,200),radius=100,color=(255,0,0),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(600,600),radius=100,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

Drawing a line 

```python
cv2.line(blank_img,pt1=(0,0),pt2=(720,720),color=(120,225,225),thickness=7)
plt.imshow(blank_img)
```
